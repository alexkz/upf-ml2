```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
%matplotlib inline
from PIL import Image
```

```python
feat_path = '/home/b.mqua/datasets/gravity/data/tutti/images/tutti_images_googlenet/pool5_7xhead t7_s1_norm.npz' 
img_files = '/home/b.mqua/datasets/gravity/data/tutti/images/tutti_images_resized.txt'
```

```python
img_paths = [x.split()[0] for x in open(img_files, 'r')]
ids = [int(re.findall('\d+',x)[0]) for x in img_paths]
pathmap = pd.Series(data=img_paths, index=ids)
idmap = pd.Series(data=np.arange(len(ids)), index=ids)
```

```python
# WARNING: REQUIRES TONS OF MEMORY 
pf = np.load(feat_path).items()[0][1]
```

```python
pf.shape
```

```python
pf_sample = pf.squeeze()
```

```python
pf_sample.shape
```

```python
pf_sim = np.dot(pf_sample, pf_sample.T)
```

```python
np.fill_diagonal(pf_sim, 0.)
```

```python
pf_sim.min(), pf_sim.max()
```

```python
most_similar = lambda x, n=10: np.argsort(pf_sim[x])[-n:][::-1]
```

```python
def show(query, img_scores):
    plt.figure()
    n = len(img_scores)
    plt.subplot(1,n+1,1)
    plt.title('QUERY')
    plt.imshow(Image.open(img_paths[query]))
    plt.axis('off')
    for i in range(n):
        idx, score = img_scores[i]
        plt.subplot(1,n+1,i+2)
        plt.title("{:.2f}".format(score))
        plt.imshow(Image.open(img_paths[idx]))
        plt.axis('off')
```

```python
query = 111
ms_imgs = most_similar(query, n=10)
ms_scores = pf_sim[query, ms_imgs]
img_scores = zip(ms_imgs, ms_scores)
show(query, img_scores)
```
